```python
def dist(x1,y1,x2,y2):
    return math.sqrt((x1-x2)**2 + (y1-y2)**2)

def intersection_point(X1,Y1,X2,Y2):
    min_dist=1e10
    ans_i1=0
    ans_i2=0
    for i1 in range(0,len(X1)):
        for i2 in range(0,len(X2)):
            if dist(X1[i1],Y1[i1],X2[i2],Y2[i2])<min_dist:
                min_dist = dist(X1[i1],Y1[i1],X2[i2],Y2[i2])
                ans_i1=i1
                ans_i2=i2
    return (X1[ans_i1]+X2[ans_i2])/2, (Y1[ans_i1] + Y2[ans_i2])/2
```

```python
def splitJacobian(J):
    T = np.concatenate([
        np.concatenate([J[0:8,0:8],J[0:8,12:18]],axis=1),
        np.concatenate([J[12:18,0:8],J[12:18,12:18]],axis=1)]
                         ,axis=0)

    # T = np.concatenate([
    #     np.concatenate([J[0:9,0:9],J[0:9,12:19]],axis=1),
    #     np.concatenate([J[12:19,0:9],J[12:19,12:19]],axis=1)]
    #                      ,axis=0)

    D = np.concatenate([
    np.concatenate([J[9:12,9:12],J[9:12,19:22]],axis=1),
    np.concatenate([J[19:22,9:12],J[19:22,19:22]],axis=1)]
                     ,axis=0)

    # D = np.concatenate([
    # np.concatenate([J[8:12,8:12],J[8:12,18:22]],axis=1),
    # np.concatenate([J[18:22,8:12],J[18:22,18:22]],axis=1)]
    #                  ,axis=0)
    return T,D



def smallestEigenVal(J):
    eigen_vals = np.linalg.eig(J)
    real_parts = [z.real for z in eigen_vals[0]]
    return min(real_parts)
```

```python
#Analyze Eigen values with CPF
t,d1,d2,td1,td2 = TDSystem5(d1CorrectionFactor=1.25, d2CorrectionFactor=2.5)
X1,Y1,J1_list = runCPF(td1,5)
X2,Y2,J2_list = runCPF(td2,5)
X3,Y3,J3_list = runCPF(t,5)

det1=[]
det_t1=[]
det_d1=[]
eig1=[]
eig_t1=[]
eig_d1=[]

for J in J1_list:
    J_T,J_D = splitJacobian(J) 
    det1.append(np.linalg.det(J))
    det_t1.append(np.linalg.det(J_T))
    det_d1.append(np.linalg.det(J_D))
    eig1.append(smallestEigenVal(J))
    eig_t1.append(smallestEigenVal(J_T))
    eig_d1.append(smallestEigenVal(J_D))   

det2=[]
det_t2=[]
det_d2=[]
eig2=[]
eig_t2=[]
eig_d2=[]

for J in J2_list:
    J_T,J_D = splitJacobian(J) 
    det2.append(np.linalg.det(J))
    det_t2.append(np.linalg.det(J_T))
    det_d2.append(np.linalg.det(J_D))
    eig2.append(smallestEigenVal(J))
    eig_t2.append(smallestEigenVal(J_T))
    eig_d2.append(smallestEigenVal(J_D))   


det3=[]
eig3=[]

for J in J3_list:
    det3.append(np.linalg.det(J))
    eig3.append(smallestEigenVal(J))
```

```python
def TDSystem4_new(requiredBuses=10, d1CorrectionFactor=1, d2CorrectionFactor=1, powerCorrectionFactor=1):
    busToAttach = 5
    trans = case9()
    # trans['bus'][busToAttach - 1][2] = 10 * 9
    # trans['bus'][busToAttach - 1][3] = 10 * 4.5 * 0.67
    trans1 = case9()
    trans2 = case9()
    
    distr1, distr2 = distrSystem4(d1CorrectionFactor=d1CorrectionFactor, d2CorrectionFactor=d2CorrectionFactor, powerCorrectionFactor=powerCorrectionFactor)
    distr3, _ = distrSystem4(d1CorrectionFactor=d1CorrectionFactor, d2CorrectionFactor=d2CorrectionFactor, powerCorrectionFactor=powerCorrectionFactor)
    distr4, _ = distrSystem4(d1CorrectionFactor=d1CorrectionFactor, d2CorrectionFactor=d2CorrectionFactor, powerCorrectionFactor=powerCorrectionFactor)

    transformer_bus = [10, 1, 0, 0, 0, 0, 1, 1, 0, 7.2, 1, 2, 0]
    transformer_branch = [[busToAttach, 10, 0.01, 0.06, 0, 500, 500, 500, 1, 0, 1, -360, 360]]
    
    trans1['bus'] = np.concatenate((trans1['bus'], [transformer_bus]))
    trans2['bus'] = np.concatenate((trans2['bus'], [transformer_bus]))
    trans1['branch'] = np.concatenate((trans1['branch'], transformer_branch))
    trans2['branch'] = np.concatenate((trans2['branch'], transformer_branch))
    
    # Reset load values on bus 5 in trans1 and trans2
    # trans1['bus'][busToAttach - 1][2] = 0
    # trans1['bus'][busToAttach - 1][3] = 0
    # trans2['bus'][busToAttach - 1][2] = 0
    # trans2['bus'][busToAttach - 1][3] = 0
    
    # Initialize distribution bus numbering starting from 11
    busNo = 11
    busNo1 = 5
    systemsAdded = 0
    distr1['bus'][0][0] = 10
    # Add distribution buses to trans1 and trans2, limit the additions
    while systemsAdded < requiredBuses:
        # Assign unique IDs to each bus in distr1 and distr2 for this system
        distr1['bus'][1][0] = busNo
        distr2['bus'][1][0] = busNo
        busNo += 1
        distr1['bus'][2][0] = busNo
        distr2['bus'][2][0] = busNo
        busNo += 1
        distr1['bus'][3][0] = busNo
        distr2['bus'][3][0] = busNo
        busNo += 1
        
        if(systemsAdded < 9):
            distr3['bus'][1][0] = busNo1
            busNo1 += 1
            distr3['bus'][2][0] = busNo1
            busNo1 += 1
            distr3['bus'][3][0] = busNo1
            busNo1 += 1
            distr4['bus'] = np.concatenate((distr4['bus'], distr3['bus'][1:]))
            distr3['branch'][0][0] = 1
            distr3['branch'][0][1] = distr3['bus'][1][0]
            distr3['branch'][1][0] = distr3['bus'][1][0]
            distr3['branch'][1][1] = distr3['bus'][2][0]
            distr3['branch'][2][0] = distr3['bus'][2][0]
            distr3['branch'][2][1] = distr3['bus'][3][0]
            distr4['branch'] = np.concatenate((distr4['branch'], distr3['branch']))


        # Add these buses and branches to trans1 and trans2
        trans1['bus'] = np.concatenate((trans1['bus'], distr1['bus'][1:]))
        trans2['bus'] = np.concatenate((trans2['bus'], distr2['bus'][1:]))
        
        distr1['branch'][0][0] = 10
        distr1['branch'][0][1] = distr1['bus'][1][0]
        distr1['branch'][1][0] = distr1['bus'][1][0]
        distr1['branch'][1][1] = distr1['bus'][2][0]
        distr1['branch'][2][0] = distr1['bus'][2][0]
        distr1['branch'][2][1] = distr1['bus'][3][0]
        trans1['branch'] = np.concatenate((trans1['branch'], distr1['branch']))
        
        distr2['branch'][0][0] = 10
        distr2['branch'][0][1] = distr2['bus'][1][0]
        distr2['branch'][1][0] = distr2['bus'][1][0]
        distr2['branch'][1][1] = distr2['bus'][2][0]
        distr2['branch'][2][0] = distr2['bus'][2][0]
        distr2['branch'][2][1] = distr2['bus'][3][0]
        trans2['branch'] = np.concatenate((trans2['branch'], distr2['branch']))
        
        systemsAdded += 1

    return trans, distr1, distr2, trans1, trans2
```

```python

def lump_all_distribution(nBlocks=10,
                          d1CorrectionFactor=1, d2CorrectionFactor=1,
                          powerCorrectionFactor=1.0):
    # 1) Start from IEEE-9
    ppc = case9()

    # 2) Remove original Pd/Qd at bus 5 (we will re-add lumped)
    ppc["bus"][4, 2:4] = 0.0

    # 3) Build & summarize all distr blocks
    P_total = 0.0
    Q_total = 0.0
    P_loss_total = 0.0
    Q_loss_total = 0.0

    for _ in range(nBlocks):
        c1, c2 = distrSystem4(d1CorrectionFactor=d1CorrectionFactor,
                              d2CorrectionFactor=d2CorrectionFactor,
                              powerCorrectionFactor=powerCorrectionFactor)
        # each block uses c1s bus loads
        P_blk, Q_blk, P_loss, Q_loss = (
            c1["bus"][:,2].sum(),
            c1["bus"][:,3].sum(),
            0.0,  # if you have loss calc, put here
            0.0
        )
        P_total      += P_blk
        Q_total      += Q_blk
        P_loss_total += P_loss
        Q_loss_total += Q_loss

    # 4) Compute equivalent series impedance so that I^2R = total losses
    #    I = (P^2+Q^2)/V  ; we take V=1.0 pu
    V0 = 1.0
    I = np.hypot(P_total, Q_total) / V0
    R_eq = P_loss_total / I**2 if I!=0 else 0.0
    X_eq = Q_loss_total / I**2 if I!=0 else 0.0

    # 5) Create one new lump bus
    new_bus = int(ppc["bus"][:,0].max() + 1)
    lump_row = np.array([ new_bus, 1, P_total, Q_total, 0,0, 1,1, 0, 345, 1, 1.1, 0.9 ])
    ppc["bus"] = np.vstack((ppc["bus"], lump_row))

    # 6) Attach series impedance from bus 5  lump_bus
    branch_row = np.array([ 5, new_bus, R_eq, X_eq, 0, 500,500,500, 0,0,1,-360,360 ])
    ppc["branch"] = np.vstack((ppc["branch"], branch_row))

    return ppc

#  Demo 
if __name__=="__main__":
    lumped_ppc = lump_all_distribution(nBlocks=10)
    print("Lumped bus (last row):", lumped_ppc["bus"][-1, :4])
    print("New branch (last row):", lumped_ppc["branch"][-1, :5])
    print(lumped_ppc)
```

```python
# Modified CPF that tracks VSI and Jacobian determinant
def runCPF_with_analysis(casedata, busForCPF, step1=0.1, step2=0.005, step3=0.01):
    X, Y, J_list = runCPF(casedata, busForCPF, step1, step2, step3)
    
    VSIs = []
    dets = []
    lambda_vals = []
    
    base_case = copy.deepcopy(casedata)
    
    for idx, J in enumerate(J_list):
        # Get current lambda value
        lambda_val = X[idx] + 1  # X stores lambda-1 values
        
        # Update case with current loading
        updated_case = copy.deepcopy(base_case)
        for i in range(len(updated_case['bus'])):
            updated_case['bus'][i][2] *= (1 + lambda_val)
            updated_case['bus'][i][3] *= (1 + lambda_val)
        
        # Calculate VSI for critical bus
        Z_th = calculate_thevenin_impedance(updated_case, busForCPF-1)
        Z_load = calculate_load_impedance(updated_case, busForCPF-1)
        VSI = abs(Z_th)/abs(Z_load) if Z_load != 0 else float('inf')
        VSIs.append(VSI)
        
        # Calculate Jacobian determinant
        det = np.linalg.det(J)
        dets.append(abs(det))
        lambda_vals.append(lambda_val)
        
    return lambda_vals, VSIs, dets

# Example usage with IEEE 9-bus system
t,_,_,_,td1,td2 = TDSystem4(d1CorrectionFactor=1, d2CorrectionFactor=1)
bus_for_analysis = 5  # Choose a PQ bus

# Run CPF with analysis
lambda_vals, VSIs, dets = runCPF_with_analysis(td1, bus_for_analysis)

# Plot results
plt.figure(figsize=(12, 6))

# Plot VSI vs Lambda
plt.subplot(1, 2, 1)
plt.plot(lambda_vals, VSIs, 'b-')
plt.axhline(1, color='r', linestyle='--')
plt.xlabel('Load Multiplier ()')
plt.ylabel('VSI')
plt.title('Voltage Stability Index vs Load')

# Plot Jacobian Determinant vs Lambda
plt.subplot(1, 2, 2)
plt.semilogy(lambda_vals, dets, 'g-')
plt.xlabel('Load Multiplier ()')
plt.ylabel('|Jacobian Determinant| (log scale)')
plt.title('Jacobian Singularity vs Load')

plt.tight_layout()
plt.show()
```
